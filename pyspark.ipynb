```python
from pyspark.sql import SparkSession
from pyspark.sql.functions import col

# Step 1: Initialize SparkSession
spark = SparkSession.builder.appName("DataFramePractice").getOrCreate()

# Step 2: Create sample data
data = [
    (1, "Ajay", 25, "IT"),
    (2, "Singh", 30, "HR"),
    (3, "Priya", 28, "Finance"),
    (4, "Anil", 35, "IT"),
    (5, "Kiran", 22, "HR")
]

# Step 3: Define column names
columns = ["id", "name", "age", "department"]

# Step 4: Create DataFrame
df = spark.createDataFrame(data, schema=columns)

# Step 5: Show DataFrame
df.show()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("WindowDemoLarge").getOrCreate()

data = [
    ("HR",  "Alice",     5000),
    ("HR",  "Bob",       6000),
    ("HR",  "Charlie",   7000),
    ("HR",  "Diana",     5500),
    ("HR",  "Evan",      6000),
    ("HR",  "Faith",     7000),

    ("IT",  "George",    5500),
    ("IT",  "Hannah",    6500),
    ("IT",  "Ian",       7500),
    ("IT",  "Jack",      5500),
    ("IT",  "Kathy",     6500),
    ("IT",  "Leo",       7500),

    ("Finance", "Mike",  8000),
    ("Finance", "Nina",  8200),
    ("Finance", "Oliver",7800),
    ("Finance", "Paul",  8200),
    ("Finance", "Queen", 8100),
    ("Finance", "Rose",  7800),

    ("Sales", "Sam",     4000),
    ("Sales", "Tara",    5000),
    ("Sales", "Uma",     4500),
    ("Sales", "Victor",  4000),
    ("Sales", "Wendy",   5000),
    ("Sales", "Xavier",  4500),

    ("Marketing", "Yara",3500),
    ("Marketing", "Zain",3600),
    ("Marketing", "Aria",3700),
    ("Marketing", "Ben", 3600),
    ("Marketing", "Cora",3800),
    ("Marketing", "Derek",3500),
]

df = spark.createDataFrame(data, ["department", "name", "salary"])
df.show(50, truncate=False)
```

```python
from pyspark.sql.window import Window
from pyspark.sql.functions import row_number

df_partition = Window.partitionBy("department").orderBy("salary")
df_partition


df = df.withColumn("row column", row_number().over(df_partition))
df.show()
```
