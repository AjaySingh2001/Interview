```python
import pandas as pd
```

```python
cmp_data = pd.ExcelFile("company_data.xlsx")

cmp_data.sheet_names
cmp_data.parse("Departments")
```

```python
cmp_sales = pd.ExcelFile("company_sales.xlsx")
cmp_sales.sheet_names
```

```python
# 1. Show first 5 Employees
# Input Sheet(s): Employees
# Display the first 5 rows of Employees
cmp_data.parse("Employees").loc[:5]
```

```python
# 2. List unique departments
# Input Sheet(s): Employees
# List unique department names
cmp_data.parse("Employees")["department"].unique().tolist()
```

```python
# 3. Count employees per city
# Input Sheet(s): Employees
# Count number of employees grouped by city

cmp_data.parse("Employees").groupby("city")["city"].count()
```

```python
# 4. Filter employees with salary > 80000
# Input Sheet(s): Employees
# Filter employees with salary > 80000
cmp_df = cmp_data.parse("Employees")
cmp_df.loc[cmp_df["salary"]>80000]
```

```python
# 5. Sort employees by joining_date (latest)
# Input Sheet(s): Employees
# Sort by joining_date desc and show top 12

cmp_data.parse("Employees").sort_values(by="joining_date", ascending=False)
```

```python
# 6. Average salary per department
# Input Sheet(s): Employees
# Calculate average salary per department

cmp_data.parse("Employees").groupby("department")["salary"].mean()
```

```python
# 7. Employees joined after 2020
# Input Sheet(s): Employees
# Find employees who joined after 2020

cmp_df = cmp_data.parse("Employees")
cmp_dt = pd.to_datetime(cmp_df["joining_date"])
cmp_df.loc[cmp_dt.dt.year>2020]
```

```python
# 8. Left merge Employees with Departments (sample)
# Input Sheet(s): Employees,Departments
# Left merge Employees with Departments and show sample

cmp_employee = cmp_data.parse("Employees")
cmp_department = cmp_data.parse("Departments")
cmp_employee.merge(cmp_department, how="left", on="department").head(20)
```

```python
# 9. Total revenue per product
# Input Sheet(s): Sales
# Compute total revenue per product

cmp_sales.parse("Sales").groupby("product", as_index=False)["price_per_unit"].aggregate(["sum"])
```

```python
# 10. Top employees by revenue
# Input Sheet(s): Sales
# Rank employees by total sales revenue
cmp_sales.parse("Sales").groupby("emp_id")["price_per_unit"].agg(["sum"])
```

```python
# 11. Identify negative salaries
# Input Sheet(s): Employees
# Find employees with negative salary values

cmp_df = cmp_data.parse("Employees")

cmp_df.loc[cmp_df["salary"]<0]
```

```python
# 12. Departments with missing locations
# Input Sheet(s): Departments
# Find departments with missing location

cmp_df = cmp_data.parse("Departments")
cmp_df.loc[(cmp_df["location"].isna()) | (cmp_df["location"] == "")]
```

```python
# 13. Offices referencing invalid emp_id
# Input Sheet(s): Offices
# List offices where emp_id not present in Employees


cmp_office_df = cmp_data.parse("Offices")
cmp_employee_df = cmp_data.parse("Employees")
cmp_office_df.loc[~cmp_office_df["emp_id"].isin(cmp_employee_df["emp_id"])]
```

```python
# 14. Sales with non-positive quantity or price
# Input Sheet(s): Sales
# Find sales with quantity <=0 or price_per_unit <=0

cmp_df = cmp_sales.parse("Sales")
cmp_df.loc[(cmp_df["quantity"]<=0) | (cmp_df["price_per_unit"]<=0)]
```

```python
# 15. Customers with duplicate names
# Input Sheet(s): Customers
# Identify customers that have duplicate names

cmp_df = cmp_sales.parse("Customers")
cmp_df.loc[cmp_df["name"].duplicated(keep=False)]
```

```python
# 16. Orders referencing invalid sale_id
# Input Sheet(s): Orders
# Find orders with sale_id not present in Sales

cmp_order_df = cmp_sales.parse("Orders")
cmp_sales_df = cmp_sales.parse("Sales")

cmp_order_df.loc[~cmp_order_df["sale_id"].isin(cmp_sales_df["sale_id"])]
```

```python
# 17. Monthly sales quantity trend (sample)
# Input Sheet(s): Sales
# Show monthly quantity sums (sample tail)
cmp_sales_df = cmp_sales.parse("Sales")
cmp_sales_df["sale_date"] = pd.to_datetime(cmp_sales_df["sale_date"])
cmp_sales_df.groupby(cmp_sales_df["sale_date"].dt.to_period("M"))["quantity"].sum().tail()

```

```python
# 18. Top products by quantity
# Input Sheet(s): Sales
# Show top products by quantity sold

cmp_sales_df = cmp_sales.parse("Sales").groupby("product")["quantity"].sum().head(10)
cmp_sales_df
```

```python
# 19. Sales left-join with employee names (sample)
# Input Sheet(s): Sales,Employees
# Left join Sales with Employees to include employee names

cmp_sales_df = cmp_sales.parse("Sales")
cmp_employee_df = cmp_data.parse("Employees")

pd.merge(cmp_sales_df, cmp_employee_df[["emp_id", "name"]], on="emp_id", how="left")
```

```python
# 20. Revenue per customer (top sample)
# Input Sheet(s): Orders,Sales
# Compute total revenue per customer (top 12)

cmp_sales.sheet_names

order = cmp_sales.parse("Orders")
sales = cmp_sales.parse("Sales")
sales
sales

sales["revenue"] = sales["quantity"] * sales["price_per_unit"]

result = pd.merge(order[["customer_id", "sale_id"]], sales[["revenue", "sale_id"]], on="sale_id")
result.groupby(["customer_id"], as_index=False)["revenue"].sum().sort_values("revenue", ascending=False).head(12)
```

```python
# 21. Q21: Pivot - product vs month quantity (sample)
# Input Sheet(s): Sales
# Create pivot table product vs month (sample)

sales = cmp_sales.parse("Sales")
sales["sale_date"] = pd.to_datetime(sales["sale_date"])

pivot_df = pd.pivot_table(sales, index="product", columns=sales["sale_date"].dt.to_period("M"), values="quantity", aggfunc="sum", fill_value=0)

pivot_df.sample(5)
```

```python
# 22. Q22: Detect duplicate sales rows
# Input Sheet(s): Sales
# Find duplicated sales rows


sales = cmp_sales.parse("Sales")
sales[sales.duplicated(keep=False)]
```
